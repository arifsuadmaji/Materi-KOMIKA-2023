# Import Library

```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
```

# METODE TABEL

```python
x1 = float(input('Masukan batas bawah : '))
x2 = float(input('Masukan batas atas : '))
N = int(input('Masukan batas banyak iterasi : '))

h = (x2 - x1) / N
i = 0

hasil = 0

print('=' * 35)
print('i       x       f(x)')
print('=' * 35)

iterasi = []
x_vals = []
f_vals = []

while (i <= N and x1 < x2):
    x = x1 + (i * h)
    fx1 = (x + np.exp(1) ** x)
    fx2 = ((x + 1) + (np.exp(1) ** (x + 1)))

    if fx1 == 0:
        hasil = x

    elif fx1 + fx2 < 0:
        if abs(fx1) < abs(fx2):
            hasil = x
        else:
            hasil = (x + 1)

    print(i, "    ", round(x, 2), "   ", (fx1))
    
    # Menyimpan data iterasi
    iterasi.append(i)
    x_vals.append(x)
    f_vals.append(fx1)
    i = i + 1

print('=' * 35)

# Menampilkan plot iterasi
plt.plot(iterasi, x_vals, marker='o', linestyle='-', color='green', label='x')
plt.plot(iterasi, f_vals, marker='o', linestyle='-', color='orange', label='f(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--', label='y=0')
plt.xlabel('Iterasi')
plt.ylabel('Nilai')
plt.grid()
plt.legend()
plt.title('Iterasi Pencarian Akar')
plt.show()
```

# METODE SECANT

```python
# Bagian Input
x0 = float(input('Masukkan nilai awal: '))
x1 = float(input('Masukkan nilai selanjutnya: '))
e = float(input('Error yang Diterima: '))
N = int(input('Jumlah Maksimum Langkah: '))

# Fungsi
def f(x):
    return x**3 - 5*x - 9  # Ganti dengan fungsi yang sesuai

# Inisialisasi list untuk menyimpan data iterasi
iterasi = []
x_vals = []
f_vals = []

# Metode Secant
print('='*60)
print("{:<13} {:<15} {:<15} {:<10}".format("Iterasi", "x2", "f(x2)", "e"))
print('='*60)

langkah = 1
kondisi = True
while kondisi:
    if f(x0) == f(x1):
        print('Error: Pembagian oleh nol!')
        break

    x2 = x0 - (x1 - x0) * f(x0) / (f(x1) - f(x0))
    error = abs(f(x2))
    print("{:<10} {:<15.6f} {:<15.6f} {:<15.6f}".format(langkah, x2, f(x2), error))
    
    # Menyimpan data iterasi
    iterasi.append(langkah)
    x_vals.append(x2)
    f_vals.append(f(x2))

    if error < e:
        print('\n Konvergen dengan error yang dapat diterima!')
        break
    
    x0 = x1
    x1 = x2
    langkah = langkah + 1

    if langkah > N:
        print('Tidak Konvergen!')
        break

print('='*60)
print('\n Akar yang Dicari adalah: %0.8f' % x2)
print('='*60)

# Menampilkan plot iterasi
plt.plot(iterasi, x_vals, marker='o', linestyle='-', color='green', label='x2')
plt.plot(iterasi, f_vals, marker='o', linestyle='-', color='orange', label='f(x2)')
plt.xlabel('Iterasi')
plt.ylabel('Nilai')
plt.grid()
plt.legend()
plt.title('Iterasi Metode Secant')
plt.show()
```

# METODE ROMBERG

```python
from tabulate import tabulate
```

```python
# Input section
lower_limit = float(input("Enter lower limit of integration: "))
upper_limit = float(input("Enter upper limit of integration: "))
order = int(input("Enter number max order: "))

def f(x):
    return x**2 * exp(x)

# Implementing trapezoidal method
def trapezoidal(a, b, n):
    # calculating step size
    h = (b - a) / n

    # Finding sum
    integration = f(a) + f(b)

    for i in range(1, n):
        k = a + i * h
        integration = integration + 2 * f(k)

    # Finding final integration value
    integration = integration * h / 2

    return integration

def romberg(k, a, b):
    if k == 1:
        return trapezoidal(a, b, 2**(k - 1))
    else:
        return (4**(k - 1) * romberg(k - 1, a, b) - romberg(k - 1, a, b)) / (4**(k - 1) - 1)

# Create list to store results
results = []
rom = zeros((order, order))

# Get trapezoid rule for first column
for i in range(order):
    rom[i, 0] = trapezoidal(lower_limit, upper_limit, 2**i)
    results.append([2**i, rom[i, 0]])

# Fill in the Romberg matrix
for i in range(1, order):
    for j in range(i, order):
        rom[j, i] = rom[j, i - 1] + (rom[j, i - 1] - rom[j - 1, i - 1]) / (4**i - 1)
        results[j].append(rom[j, i])

# Display results as a table

headers = ["h", *["Romberg({},{})".format(j, i) for i in range(order)]]
print(tabulate(results, headers=headers, tablefmt="fancy_grid"))

# Print the final result using Romberg integration
result = rom[order - 1, order - 1]
print("Hasil Estimasi Integral:", result)
```

# METODE EULER

```python
import matplotlib.pyplot as plt
import math

# Input Section
h = float(input("Masukkan panjang langkah (h): "))
x0 = float(input("Masukkan nilai awal x0: "))
y0 = float(input("Masukkan nilai awal y0: "))
n = int(input("Masukkan jumlah iterasi (n): "))

# Implementing Euler's Method
print("="*50)
print("{:<10} {:<20} {:<20}".format("Iterasi", "x", "y"))
print("="*50)

x = x0  # Nilai awal x
y = y0  # Nilai awal y
# Function
dy = lambda x, y: math.sin(x)**2 * y
f = lambda x: 2 * math.exp(0.5 * (x - math.sin(x) * math.cos(x)))

# Menampilkan hasil untuk iterasi pertama
print("{:<10} {:<20} {:<20}".format(0, x, y))

x_plot = []
y_euler = []
y_analytical = []

# Euler's Method
for i in range(1, n + 1):
    x_plot.append(x)
    y_euler.append(y)
    y_analytical.append(f(x))
    y = y + dy(x, y) * h
    x = x + h
    print("{:<10} {:<20} {:<20}".format(i, x, y))

x_plot.append(x)
y_euler.append(y)
y_analytical.append(f(x))

print("="*50)

# Plotting
plt.plot(x_plot, y_euler, marker='o', linestyle='-', color='green', label="Euler's Method")
plt.plot(x_plot, y_analytical, marker='o', linestyle='-', color='orange', label="Analytical Solution")
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Euler\'s Method vs Analytical Solution')
plt.show()
```

# METODE NEWTON RAPHSON

```python
# Bagian Input
x0 = float(input('Masukkan Tebakan Awal: '))
e = float(input('Error yang Diterima: '))
N = int(input('Langkah Maksimum: '))

# Definisi Fungsi
def f(x):
    return x**3 - 5*x - 9

# Definisi Turunan Fungsi
def g(x):
    return 3*x**2 - 5

# Inisialisasi
step = 1
flag = 1
condition = True
x_vals = []
f_vals = []

# Implementasi Metode Newton Raphson
print("="*70)
print("{:<12} {:<15} {:<15} {:<15}".format("Iterasi", "x1", "f(x1)", "e"))
print("="*70)

while condition:
    if g(x0) == 0.0:
        print('Error: Pembagian oleh nol!')
        break
    
    x1 = x0 - f(x0)/g(x0)
    error = abs(f(x1))
    print("{:<10} {:<15.6f} {:<15.6f} {:<15.6f}".format(step, x1, f(x1), error))
    
    # Menambahkan nilai ke dalam list untuk plotting
    x_vals.append(x1)
    f_vals.append(f(x1))
    
    x0 = x1
    step = step + 1
    
    if step > N:
        flag = 0
        break
    
    condition = error > e

print("="*70)

if flag == 1:
    print('\nAkar yang dicari adalah: %0.8f' % x1)
else:
    print('\nTidak Konvergen.')
print("="*70)

# Menampilkan plot hasil iterasi
plt.plot(range(1, step), x_vals, marker='o', linestyle='-', color='green', label='Nilai x1')
plt.plot(range(1, step), f_vals, marker='o', linestyle='-', color='orange', label='Nilai f(x1)')
plt.xlabel('Iterasi')
plt.ylabel('Nilai')
plt.grid()
plt.legend()
plt.title('Iterasi Metode Newton Raphson')
plt.show()
```
