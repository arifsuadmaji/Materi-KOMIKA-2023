## Membuat dataframe

```python
data = {'kolom1':['ayam', 'kucing', 'sapi', 'kambing'], 
        'kolom2':['1', '4', '9', '16']}

df = pd.DataFrame(data)
df
```

## Memilih kolom berdasarkan tipe data

```python
n_rows = 10
n_cols = 2
kolom = ['float', 'int']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=kolom)
df['float'] = df['float'].astype('float')

df
```

## Membuat datetime

```python
df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df = df.reset_index()
df
```

## Membuat Kolom baru
