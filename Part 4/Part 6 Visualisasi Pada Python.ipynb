# Bar Plot

Bar plot atau diagram batang adalah jenis plot yang digunakan untuk memvisualisasikan data kategorikal yang memiliki nilai numerik. Pada bar plot, sumbu horizontal biasanya menunjukkan variabel kategori, sedangkan sumbu vertikal menunjukkan nilai numerik yang terkait.

Bar plot sangat berguna untuk membandingkan nilai antara beberapa kategori. Jenis bar plot yang paling umum adalah vertical bar plot, di mana batang vertikal menunjukkan nilai pada sumbu vertikal. Namun, ada juga horizontal bar plot, di mana batang horizontal menunjukkan nilai pada sumbu vertikal.

__Contoh Kasus:__

Seorang peneliti ingin menunjukkan seberapa banyak penduduk di beberapa kota di Indonesia. Dia menggunakan bar plot untuk menampilkan data tersebut. Berikut adalah contoh kode program Python untuk membuat bar plot:

````python
# Data
kota = ['Jakarta', 'Surabaya', 'Bandung', 'Medan', 'Makassar']
penduduk = [10770487, 2928323, 2394000, 2500239, 1425000]
````

````python
# Membuat bar plot
plt.bar(kota, penduduk)

# Menampilkan plot
plt.show()
````

````python
# Membuat bar plot
plt.bar(kota, penduduk)
# Memberi judul dan label pada sumbu
plt.title('Penduduk di Beberapa Kota di Indonesia')
plt.xlabel('Kota')
plt.ylabel('Jumlah Penduduk')
# Menampilkan plot
plt.show()
````

````python
# Membuat bar plot
plt.bar(kota, penduduk, color = 'green')
# Memberi judul dan label pada sumbu
plt.title('Penduduk di Beberapa Kota di Indonesia')
plt.xlabel('Kota')
plt.ylabel('Jumlah Penduduk')
# Menampilkan plot
plt.show()
````

````python
df = pd.DataFrame({
    'Kota': kota,
    'Penduduk': penduduk,
})

# Mengurutkan data berdasarkan populasi penduduk terkecil
df = df.sort_values(by='Penduduk')

x_coords = np.arange(len(df))
colors = ['orange' for _ in range(len(df))]
colors[-1] = 'red'

# Membuat bar plot
plt.bar(x_coords, df['Penduduk'], tick_label=df['Kota'], color = colors)
# Memberi judul dan label pada sumbu
plt.title('Penduduk di Beberapa Kota di Indonesia')
plt.xlabel('Kota')
plt.ylabel('Jumlah Penduduk')
# Menampilkan plot
plt.show()
````

````python
# Membuat bar plot
plt.barh(x_coords, df['Penduduk'], tick_label=df['Kota'], color = colors)
# Memberi judul dan label pada sumbu
plt.title('Penduduk di Beberapa Kota di Indonesia')
plt.xlabel('Jumlah Penduduk')
plt.ylabel('Kota')
# Menampilkan plot
plt.show()
````

# Pie Plot

Pie plot adalah jenis plot yang menampilkan proporsi data dalam bentuk lingkaran atau pie chart. Pie plot umumnya digunakan untuk menampilkan data kategorikal atau data diskrit yang proporsinya dapat dinyatakan sebagai persentase dari keseluruhan.

Setiap proporsi data direpresentasikan sebagai bagian dari lingkaran, di mana ukuran setiap bagian proporsional dengan nilai data yang direpresentasikan. Pie plot biasanya digunakan untuk menunjukkan perbandingan antara beberapa kategori data, atau untuk menyoroti proporsi data yang lebih besar atau lebih kecil dari yang lain.

__Contoh Kasus:__

Seorang manajer proyek ingin menunjukkan seberapa besar anggaran yang telah dialokasikan untuk setiap bagian dari proyek. Berikut data yang dimiliki:

````python
# Data
bagian_proyek = ['Desain', 'Pengembangan', 'Pengujian', 'Implementasi','Pemeliharaan']
anggaran = [15000000, 30000000, 12000000, 25000000, 8000000]
````

````python
# Membuat pie plot
plt.pie(
    anggaran, 
    labels=bagian_proyek
)

# Menampilkan plot
plt.show()
````

````python
# Membuat pie plot
plt.pie(
    anggaran, 
    labels=bagian_proyek, 
    autopct='%1.1f%%'
)

# Memberi judul pada plot
plt.title('Anggaran Proyek')
# Menampilkan plot
plt.show()
````

````python
# Membuat pie plot
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
plt.pie(
    anggaran, 
    labels=bagian_proyek, 
    autopct='%1.1f%%',
    colors=warna    
)

# Memberi judul pada plot
plt.title('Anggaran Proyek')
# Menampilkan plot
plt.show()
````

````python
explode = (0, 0, 0, 0.1, 0)
# Membuat pie plot
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
plt.pie(
    anggaran, 
    labels=bagian_proyek, 
    autopct='%1.1f%%',
    colors=warna,
    explode = explode,
    shadow = True
)

# Memberi judul pada plot
plt.title('Anggaran Proyek')
# Menampilkan plot
plt.show()
````

````python
explode = (0, 0, 0, 0.1, 0)
# Membuat pie plot
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
plt.pie(
    anggaran, 
    labels=bagian_proyek, 
    autopct='%1.1f%%',
    colors=warna,
    explode = explode,
    shadow = True,
    startangle=270
)

# Memberi judul pada plot
plt.title('Anggaran Proyek')
# Menampilkan plot
plt.show()
````

## Line Plot

Line plot atau disebut juga sebagai line chart adalah sebuah jenis visualisasi data yang digunakan untuk menampilkan perubahan data dalam rentang waktu tertentu. Line plot terdiri dari sumbu-x dan sumbu-y, dimana sumbu-x mewakili waktu atau interval data dan sumbu-y mewakili data numerik yang berkaitan dengan waktu tersebut. Garis pada line plot menghubungkan titik data yang terurut berdasarkan waktu atau interval tertentu. Line plot sangat berguna dalam menunjukkan tren data seiring waktu dan perubahan data pada interval tertentu. 

__Contoh Kasus:__

Seorang peneliti ingin menunjukkan bagaimana populasi burung di suatu kawasan berubah selama beberapa tahun terakhir. Berikut data yang dimiliki:

````python
# Data
tahun = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
populasi = [1000, 1200, 1300, 1400, 1600, 1800, 2000, 2200, 2400, 2500]
````

##### Line plot sederhana

````python
# Membuat line plot
plt.plot(tahun, populasi)
# Menampilkan line plot
plt.show()
````

##### Line plot sederhana + judul & label

````python
# Membuat line plot
plt.plot(tahun, populasi)
# Memberi judul dan label pada sumbu
plt.title('Perubahan Populasi Burung')
plt.xlabel('Tahun')
plt.ylabel('Populasi')
# Menampilkan plot
plt.show()
````

# Scatter Plot

Scatter plot adalah salah satu jenis plot yang digunakan untuk menampilkan hubungan antara dua variabel numerik. Plot ini terdiri dari titik-titik yang mewakili nilai-nilai dari dua variabel tersebut, dengan sumbu x dan y yang merepresentasikan nilai dari masing-masing variabel. Selain itu, scatter plot juga dapat digunakan untuk melihat pola penyebaran pada data.

<br>

__Contoh kasus 1:__

Seorang peneliti ingin melihat apakah ada hubungan antara tinggi badan dan berat badan siswa. Berikut data yang dimiliki:

````python
# Data
x = [152, 167, 175, 162, 172, 185, 168, 177, 159, 182]
y = [50, 57, 64, 54, 62, 73, 58, 68, 49, 70]
````

````python
# Membuat scatter plot
plt.scatter(x, y)
# Menampilkan plot
plt.show()
````

````python
# Membuat scatter plot
plt.scatter(x, y, color='green')
# Memberi judul dan label pada sumbu
# Menampilkan plot
plt.show()
````

````python
# Membuat scatter plot
plt.scatter(x, y, color='green')
# Memberi judul dan label pada sumbu
plt.title('Sebaran data Berat dan Tinggi Badan')
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Berat Badan (kg)')
# Menampilkan plot
plt.show()
````

__Contoh kasus 2:__

Seorang peneliti sedang jumlah pasien yang terinfeksi virus pada wilayah yang berbeda. peniliti menemukan pasien yang menyebar secara berkelompok sebagai berikut.

````python
# Data pasien yang terinfeksi pada masing-masing wilayah
wilayah_1 = [35, 27, 10, 43, 12, 8, 23, 7, 4, 17, 21, 6, 30, 20, 14]
kepadatan_penduduk1 = [1000, 1500, 2000, 1800, 3000, 3500, 4000, 2200, 1800, 4500, 3300, 1200, 4200, 1100, 3300]
wilayah_2 = [38, 42, 18, 15, 18, 11, 22, 15, 13, 9, 24, 36, 19, 29, 7]
kepadatan_penduduk2 = [3088, 1578, 3264, 4225, 1189, 3355, 3678, 3367, 4555, 1456, 4567, 4531, 3154, 4890, 4632]
````

````python
# Memvisualisasikan data dengan scatter plot
plt.scatter(kepadatan_penduduk1, wilayah_1, color = 'red')
plt.scatter(kepadatan_penduduk1, wilayah_2, color = 'purple')
plt.title('Penyebaran Penyakit di Beberapa Wilayah')
plt.xlabel('Kepadatan penduduk')
plt.ylabel('Jumlah Pasien Terinfeksi')
plt.show()
````

# Heatmap

Heatmap adalah sebuah visualisasi yang digunakan untuk menganalisis kumpulan data numerik dengan memberikan warna yang berbeda-beda pada setiap kotak yang merepresentasikan nilai pada data. Dalam heatmap, data numerik diubah menjadi warna, sehingga memberikan gambaran visual tentang keterkaitan atau pola pada data. 

Heatmap biasanya digunakan untuk menganalisis hubungan antara dua variabel numerik atau lebih, sehingga kita dapat mengidentifikasi pola atau tren pada data tersebut. Heatmap juga dapat digunakan untuk mengidentifikasi titik-titik yang memiliki nilai ekstrem atau anomali pada data.


__Contoh kasus:__

Kita memiliki nilai 20 mahasiswa dari jurusan yang berbeda-beda. Setiap baris data mewakili jurusan dari setiap mahasiswa. Sedangkan, setiap kolom mewakili nilai dari masing-masing mahasiswa, mulai dari mahasiswa pertama (1) sampai mahasiswa ke-20.

````python
# Data
nilai_mahasiswa = [
    [80, 69, 72, 86, 56, 94, 69, 55, 61, 56, 56, 73, 90, 92, 92, 87, 83, 54, 79, 53], # Matematika
    [82, 68, 59, 95, 72, 99, 88, 87, 54, 80, 87, 67, 88, 63, 99, 81, 59, 75, 76, 59], # Fisika
    [61, 60, 82, 93, 73, 64, 59, 86, 55, 95, 53, 89, 93, 51, 51, 97, 86, 58, 99, 53], # Kimia
    [83, 62, 93, 79, 59, 87, 62, 52, 64, 55, 61, 96, 79, 86, 74, 65, 82, 61, 95, 84], # Biologi
    [53, 77, 63, 61, 61, 97, 61, 84, 77, 95, 97, 79, 68, 57, 71, 73, 50, 54, 53, 67], # Bahasa Inggris
]
    
jurusan = ['Matematika', 'Fisika', 'Kimia', 'Biologi', 'Bahasa Inggris']
````

````python
# membuat heatmap
mahasiswa = [i for i in range(1, 21)]
sns.heatmap(nilai_mahasiswa, yticklabels=jurusan, xticklabels=mahasiswa)
plt.show()
````

````python
# membuat heatmap
mahasiswa = [i for i in range(1, 21)]
plt.figure(figsize=(6,4))
sns.heatmap(nilai_mahasiswa, yticklabels=jurusan, xticklabels=mahasiswa, cmap='RdYlGn')
plt.show()
````

````python
# membuat heatmap
mahasiswa = [i for i in range(1, 21)]
plt.figure(figsize=(6,4))
sns.heatmap(nilai_mahasiswa, yticklabels=jurusan, xticklabels=mahasiswa, cmap='RdYlGn')
plt.xlabel('Mahasiswa')
plt.ylabel('jurusan')
plt.title('Nilai Mahasiswa pada Setiap Jurusan')
plt.show()
````
